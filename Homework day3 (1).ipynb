```python
#User Login Application
#        - get Username and Password values from user.
#        - Check the values in an if statment and tell the user if they were successfull


userinput = input("Please Input ID:")

passinput = input("Please Input Pass:")

if userinput == "user1" and passinput== "pass1":

    print("well done")

else:
    print("id or pass wrong")
```
